```python
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\Gangaa\Downloads\synthetic_climate_data.csv")
data
```

```python
imputer = SimpleImputer(strategy='median')
data[['Rainfall (mm)', 'Temperature (°C)']] = imputer.fit_transform(data[['Rainfall (mm)', 'Temperature (°C)']])
```

```python
fig2, ax2 = plt.subplots(figsize=(8, 6))
sns.boxplot(data=data, x='Soil Quality', y='Rainfall (mm)', ax=ax2, palette='muted')
ax2.set_title('Rainfall Distribution by Soil Quality')
fig2.show()
```

```python
fig3, ax3 = plt.subplots(figsize=(8, 6))
sns.histplot(data['Temperature (°C)'], kde=True, bins=30, ax=ax3, color='skyblue')
ax3.set_title('Distribution of Temperature')
fig3.show()
```

```python
fig4, ax4 = plt.subplots(figsize=(8, 6))
sns.countplot(data=data, x='Crop Suitability', order=data['Crop Suitability'].value_counts().index, ax=ax4, palette='pastel')
ax4.set_title('Count of Crop Suitability Types')
fig4.show()
```

```python
fig6, ax6 = plt.subplots(figsize=(8, 6))
sns.violinplot(data=data, x='Historical Yield Trend', y='Temperature (°C)', ax=ax6, palette='Set2')
ax6.set_title('Temperature Distribution by Historical Yield Trend')
fig6.show()
```

```python
fig1, ax1 = plt.subplots(figsize=(8, 6))
sns.scatterplot(data=data, x='Temperature (°C)', y='Rainfall (mm)', hue='Soil Quality', ax=ax1, palette='viridis')
ax1.set_title('Temperature vs Rainfall (Colored by Soil Quality)')
fig1.show()
```

```python
fig5, ax5 = plt.subplots(figsize=(8, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', ax=ax5)
ax5.set_title('Correlation Heatmap')
fig5.show()
```

```python
fig7 = sns.pairplot(data, vars=['Latitude', 'Longitude', 'Temperature (°C)', 'Rainfall (mm)'], hue='Soil Quality', palette='husl')
fig7.set_title('Average Rainfall by Soil Quality')
```

```python
fig8, ax8 = plt.subplots(figsize=(8, 6))
avg_rainfall = data.groupby('Soil Quality')['Rainfall (mm)'].mean().reset_index()
sns.lineplot(data=avg_rainfall, x='Soil Quality', y='Rainfall (mm)', marker='o', ax=ax8, color='red')
ax8.set_title('Average Rainfall by Soil Quality')
fig8.show()
```

```python
fig9 = sns.jointplot(data=data, x='Temperature (°C)', y='Rainfall (mm)', kind='hex', cmap='Blues')
fig9.set_title('Average Rainfall by Soil Quality')
```

```python
fig10, ax10 = plt.subplots(figsize=(8, 6))
sns.stripplot(data=data, x='Crop Suitability', y='Rainfall (mm)', ax=ax10, palette='deep')
ax10.set_title('Rainfall by Crop Suitability')
fig10.show()
```
